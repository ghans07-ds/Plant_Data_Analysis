```python
plants_file=[]
mainFile=pd.DataFrame()
for i in range(len(list_plants)):
    file=pd.read_csv("E://DATA Science//komal//f_done//"+list_plants[i]+".csv",encoding='latin1')
    if i==0:
        mainFile=file
    else:
        mainFile=pd.concat([mainFile,file])
```
